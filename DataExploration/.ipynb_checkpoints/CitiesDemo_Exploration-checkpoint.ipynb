# Data Exploration for us cities demographic data

```python
import pandas as pd
```

```python
raw_citydemo = pd.read_csv('us-cities-demographics.csv',sep = ';')
```

```python
#what does it look like?
raw_citydemo.head()
```

```python
#how big is it
raw_citydemo.shape
```

```python
#the race column is weird, does each citi has more than 1 row?
raw_citydemo.groupby(['City']).count().head()
```

```python
#why Albany has 10 row?
raw_citydemo[raw_citydemo['City']=='Albany']
```

```python
#Citi and State together should be the primary key
raw_citydemo.groupby(['City','State']).count().head()
```

```python
#these are the columns that don't need any changes, but needs to dedupe
citydemo_unchanged = raw_citydemo.iloc[:,0:10].drop_duplicates()
citydemo_unchanged.head()
```

```python
#these are columns that are stacked
citydemo_stacked = raw_citydemo.iloc[:,[0,1,10,11]]
citydemo_stacked.head()
```

```python
#make sure the naming used for Race are consistent
citydemo_stacked.groupby('Race').count()
```

```python
#get them a shorter name - no offense, all people on earth deserves our repect equally
citydemo_stacked['Race'] = citydemo_stacked['Race'].replace(['American Indian and Alaska Native',
                                               'Black or African-American',
                                               'Hispanic or Latino'],
                                              ['Native','Black','Hispa'])
```

```python
citydemo_stacked.groupby('Race').count()
```

```python
#unstacking
citydemo_unstacked = citydemo_stacked.pivot_table(index=['City','State'], columns="Race", values="Count", aggfunc= 'first').reset_index().rename_axis(None, 1)
```

```python
citydemo_unstacked.head()
```

```python
# see the count of the 2 parts of dataset we want to merge
print(citydemo_unchanged.shape)
print(citydemo_unstacked.shape)
```

```python
citydemo = citydemo_unchanged.merge(citydemo_unstacked,on = ['City','State'])
```

```python
citydemo.shape
```

```python
citydemo.head()
```

```python
citydemo.columns = citydemo.columns.str.replace(' ', '')
```

```python
citydemo.columns = citydemo.columns.str.replace('-', '')
```

```python
citydemo.head()
```

```python
citydemo.to_csv('dataoutput/citydemo.csv',index=False)
```
