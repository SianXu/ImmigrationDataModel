```python
# Do all imports and installs here
import pandas as pd
from datapackage import Package
```

```python
data_url = 'https://datahub.io/core/airport-codes/datapackage.json'

# to load Data Package into storage
package = Package(data_url)

# to load only tabular data
resources = package.resources
for resource in resources:
    print(resource.descriptor['path'])
```

```python
data1 = pd.read_csv(package.resources[1].descriptor['path'])
data1.shape
```

```python
data5 = pd.read_csv(package.resources[5].descriptor['path'])
data5.shape
```

different resource doesn't seem to have much difference, i'll use the _csv

```python
#let's see what it looks like
raw_airportcode = data1
raw_airportcode.head()
```

```python
#see how many are missing
raw_airportcode.isnull().sum()
```

```python
# why would municipality be missing? how do we know which city is it then?
raw_airportcode[raw_airportcode['municipality'].isnull()].head()
```

```python
#is ident unique??
raw_airportcode['ident'].is_unique
```

```python
#see countries with most airports
raw_airportcode.iloc[:,5:7].groupby('iso_country').count().sort_values(by = 'iso_region',ascending = False).head(10)
```

```python
#what iso_region is like for non US airports??
raw_airportcode[raw_airportcode['iso_country']=='CA'].head()
```

```python
#check Chinese airports as i'm more familiar witht the names
raw_airportcode[raw_airportcode['iso_country']=='CN'].head()
```

```python
#I don't see much changes we need
airportcode = raw_airportcode
```

```python
airportcode.head()
```

```python
# create a separate region column
airportcode = airportcode.assign(iso_region2 = lambda x: x['iso_region'].str.split('-',1).str[-1])
```

```python
# create separate lat log
airportcode[['lat','lon']] = airportcode['coordinates'].str.split(',',1,expand = True)
```

```python
airportcode.head()
```

```python
airportcode.to_csv('dataoutput/airportcode.csv',index = False)
```
